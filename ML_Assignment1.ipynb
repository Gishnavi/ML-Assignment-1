```python
import numpy as np

# create random vector of size 15 having only Integers in the range 1-20.
vtr= np.random.randint(1, 20, size=15)

# Reshape the array to 3 by 5
reshape_array = vtr.reshape(3, 5)

# Print array shape
print("THE SHAPE OF THE ARRAY IS:", reshape_array.shape)

# Replace the max in each row by 0
reshape_array[np.arange(3), np.argmax(reshape_array, axis=1)] = 0

print("THE NEW ARRAY IS :")
print(reshape_array)

# Create a 2-dimensional array 
arr = np.zeros((4, 3), dtype=np.int32)

# Print shape, type, and data type of the array
print("THE SHAPE OF THE ARRAY IS:", arr.shape)
print("TTHE TYPE OF THE ARRAY IS:", type(arr))
print("THE DATA TYPE OF THE ARRAY IS:", arr.dtype)
```

```python
import numpy as np

# defining an array
sq_arr = np.array([[3, -2], [1, 0]])
print("\nThe Array is:",sq_arr)

# Computing the eigenvalues and right eigenvectors
eigen_val,r_eigen_vtr = np.linalg.eig(sq_array)

print("\nRight Eigenvectors:")
for eigenvector in r_eigen_vtr:
    print(eigenvector)   
print("\nEigenvalues:")
for eigenvalue in eigen_val:
    print(eigenvalue)


```

```python
import matplotlib.pyplot as plt

programming_lang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# index of the slice containing the highest value.
index_max = popularity.index(max(popularity))

explode = [0] * len(programming_lang)  
#Assign an explode value of 0.1 to the slice that has the highest value.
explode[index_max] = 0.1 

# Change the initial position of the pie chart slices.
startangle = 90 - (360 *index_max) / len(programming_lang) + 45

plt.pie(popularity, labels=programming_lang, autopct='%1.1f%%', explode=explode, shadow=True, startangle=startangle)
plt.title('Popularity of Programming Languages')

plt.axis('equal') 

plt.show()
```
